Exercise 2: Factorial of a Number


```python
num = int(input("Enter a number: "))
factorial = 1

for i in range(1, num + 1):
    factorial *= i

print("Factorial:", factorial)
```

Exercise 4: Simple Calculator Using Functions

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b

x = float(input("Enter first number: "))
y = float(input("Enter second number: "))

print(add(x, y))
print(subtract(x, y))
print(multiply(x, y))
print(divide(x, y))
```

Exercise 5: Remove First Duplicate from List

```python
my_list = [1, 2, 3, 2, 4, 5]

seen = set()
for item in my_list:
    if item in seen:
        my_list.remove(item)
        break
    seen.add(item)

print(my_list)
```

Exercise 6: Extract Keys from Dictionary

```python
sample_dict = {
    "name": "Anderson",
    "age": 35,
    "salary": 5000,
    "city": "London"
}

keys = ["name", "salary"]
new_dict = {}

for key in keys:
    new_dict[key] = sample_dict[key]

print(new_dict)
```

Exercise 7: Convert Two Lists into Dictionary

```python
list_one = ['one', 'two', 'three']
list_two = [1, 2, 3]

result = {}

for i in range(len(list_one)):
    result[list_one[i]] = list_two[i]

print(result)
```
